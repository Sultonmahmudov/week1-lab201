````python
study_hours = [1, 2, 3, 4, 5]
exam_scores = [50, 55, 65, 70, 80]


for i in range(len(study_hours)):
    print(f"Hours of study: {study_hours[i]} - Exam score: {exam_scores[i]}")
````

````python
study_exam_dict = {study_hours[i]: exam_scores[i] for i in range(len(study_hours))}
display(study_exam_dict)
````

````python
study_exam_dict.keys()
````

#problem 6

````python
scores = [35, 58, 62, 74, 89, 45]

def categories(score):
  if score < 50 :
    return "low"
  elif score > 50 and score < 70 :
    return "medium"
  else:
    return "high"

for score in scores:
  result = categories(score)
  print(f"Input {score} -- output {result}")
````
